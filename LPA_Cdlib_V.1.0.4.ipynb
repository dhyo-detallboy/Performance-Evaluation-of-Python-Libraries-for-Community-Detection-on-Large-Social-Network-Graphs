### Import Library

```python
import pandas as pd
import time 
import textwrap
from cdlib import algorithms
import networkx as nx
import psutil

from cdlib import evaluation
```

### Fungsi Untuk Mengukur Runtime & Memory Usage

```python
# Fungsi untuk menghitung Runtime
def print_runtime(dataset, total_runtime):
    hours, remainder = divmod(total_runtime, 3600)
    minutes, seconds = divmod(remainder, 60)
    print(f"{dataset.ljust(13)}: {int(hours)} jam, {int(minutes)} menit, {int(seconds)} detik")

# Fungsi untuk menghitung Memory Usage
def get_memory():
    process = psutil.Process()
    return process.memory_info().rss / (1024 ** 2)  # Konversi ke MB
```

```python
# Mendefinisikan total Runtime
algorithm_runtime = 0

# Mendefinisikan total Memory Usage
algorithm_memory_usage = 0

```

### Menghitung Runtime & Memory All (Start)

```python
# Mendapatkan Runtime awal (Program)
all_runtime_start = time.time()

# Mendapatkan Memory Usage sebelum eksekusi (Program)
all_memory_before = get_memory()
```

### Import Dataset

```python
path_df =r"D:\KULIAH\Semester 8\Source Code TA\TA Version 1.0.4 V2\Used_Dataset\com-amazon.ungraph.txt\com-amazon.ungraph.txt"

df = pd.read_csv(path_df, delimiter='\s+', header=None, names=['from_node', 'to_node'])
```

### Pemodelan Graf Dataset

```python
# Convert DataFrame to edge list
edge_list = [tuple(x) for x in df.to_numpy()]
```

```python
G = nx.from_edgelist(edge_list)
```

```python
# Menampilkan jumlah node dan edge
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()

print(f"Jumlah node: {num_nodes}")
print(f"Jumlah edge: {num_edges}")
```

### Pengecekan Jenis Graf Dataset

```python
# Memeriksa jenis graf directed atau undirected
graph_type = "Directed" if G.is_directed() else "Undirected"
print("Graph type:", graph_type)
```

### Implementasi Deteksi Komunitas

### Label Propagation

```python
# Mendapatkan Timestamp Awal
start_runtime_lpa = time.time()

# Mendapatkan Runtime awal (Program)
lpa_memory_start = get_memory()

# Implementasi Algoritma Label Propagation
coms_label_propagation = algorithms.label_propagation(G)

# Menghitung jumlah komunitas yang ditemukan
num_communities_label_propagation = len(coms_label_propagation.communities)
print(f"Clustering with {G.number_of_nodes()} elements and {num_communities_label_propagation} clusters")

print()

# Menampilkan hasil deteksi komunitas
for idx, com in enumerate(coms_label_propagation.communities):
    print(f"Community {idx + 1}: {com}")

print()

# Menghitung Modularity Density
modularity = evaluation.erdos_renyi_modularity(G, coms_label_propagation)
print(f"Modularity : {modularity.score}")

# Mendapatkan Timestamp Akhir
end_runtime_lpa = time.time()

# Mendapatkan Runtime awal (Program)
lpa_memory_end = get_memory()
```

### Menghitung All Runtime & Memory (End)

```python
# Mendapatkan timestamp akhir (Program)
all_runtime_end = time.time()

# Mendapatkan Memory Usage setelah eksekusi (Program)
all_memory_after = get_memory()
```

### Hasil Perhitungan All Runtime & Memory

```python
# Menghitung Total Runtime (Program)
all_runtime = all_runtime_end - all_runtime_start

# Menghitung Total Memory Usage (Program)
all_memory_used = all_memory_after - all_memory_before

# Menghitung Total Runtime (Program)
lpa_runtime = end_runtime_lpa - start_runtime_lpa

# Menghitung Total Memory Usage (Program)
lpa_memory_used = lpa_memory_end - lpa_memory_start
```

### Rekapitulasi Hasil

```python
print("+------------------------------------------------------+")
print("+----------+ Rekapitulasi Runtime & Memory +-----------+")
print("+------------------------------------------------------+")
print("Runtime Program (Seconds)   :", all_runtime, "detik")
print("Memory Program (MB)         :", all_memory_used, "MB")
print("+------------------------------------------------------+")
print("Runtime Algoritma (Seconds) :", lpa_runtime, "detik")
print("Memory Usage (MB)           :", lpa_memory_used, "MB")
print("+------------------------------------------------------+")
print("Modularity                  :", modularity.score)
print("+------------------------------------------------------+")

```
